```python
df = newdf
df['postedOn']= pd.to_datetime(df['postedOn'])
df['date_posted'] = [d.date() for d in df['postedOn']]
df['time_posted'] = [d.hour for d in df['postedOn']]


days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
df['day_posted'] = [d.weekday() for d in df.postedOn]
df=df.sort_values(by='day_posted')
df['day_posted'] = [days[d] for d in df.day_posted]
df=df.reset_index(drop=True)
#df=df.drop('postedOn',axis=1)

df.head()
```

```python
fig = plt.figure(figsize=(12,12),facecolor=BgCol)
#++++++++++++++++++++++++++++++++++++++AX3 -> LIKES PER DAY
likes_by_day = df.copy()
likes_by_day['day_posted'] = pd.Categorical(likes_by_day['day_posted'], categories=
    ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday', 'Sunday'],
    ordered=True)

likes_by_day['day_posted'] = likes_by_day.day_posted.map({
    'Monday':'Mon',
    'Tuesday':'Tues',
    'Wednesday':'Weds',
    'Thursday':'Thurs',
    'Friday':'Fri',
    'Saturday':'Sat',
    'Sunday':'Sun'
})

likes_by_day=likes_by_day[['likes','day_posted']].groupby('day_posted').agg('sum').reset_index(drop=False)

posts_per_day = df.copy()

posts_per_day['day_posted'] = posts_per_day.day_posted.map({
    'Monday':'Mon',
    'Tuesday':'Tues',
    'Wednesday':'Weds',
    'Thursday':'Thurs',
    'Friday':'Fri',
    'Saturday':'Sat',
    'Sunday':'Sun'
})
days=['Mon','Tues','Weds','Thurs','Fri','Sat', 'Sun']



posts_per_day=posts_per_day['day_posted'].value_counts().rename_axis('Day').to_frame('counts').reindex(index=days).reset_index()

likes_by_day = likes_by_day.merge(posts_per_day,how='left',left_on='day_posted',right_on='Day').drop('Day',axis=1)



likes_by_day['likesPerPost'] = likes_by_day.apply(lambda x: x['likes']/x['counts'], axis=1)


plt.show()
```

```python
likes_by_day = df.copy()

likesMean = likes_by_day[['likes','day_posted']].groupby(['day_posted']).agg({'likes':['mean','std']})

likesMean.columns = likesMean.columns.get_level_values(0)
likesMean.columns = ['Mean Likes','Stdev']
likesMean=likesMean.reset_index()
likesMean['day_posted'] = likesMean.day_posted.map({
    'Monday':'Mon',
    'Tuesday':'Tues',
    'Wednesday':'Weds',
    'Thursday':'Thurs',
    'Friday':'Fri',
    'Saturday':'Sat',
    'Sunday':'Sun'
})
likesMean = likesMean.merge(posts_per_day,how='left',left_on='day_posted',right_on='Day').drop('Day',axis=1)
days=['Mon','Tues','Weds','Thurs','Fri','Sat', 'Sun']

likesMean['likesPerPost'] = likesMean.apply(lambda x: x['Mean Likes']/x['counts'], axis=1)
likesMean = likesMean.set_index('day_posted').reindex(index=days).reset_index()



likesMean
```
